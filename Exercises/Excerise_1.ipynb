```python
from typing import TypedDict,Dict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    compliment: str



def complement_node(state: AgentState) -> AgentState:

    """Complements the user"""

    state['compliment'] = state['compliment'] + ", you are doing an amazing job learning LangGraph. You can do this"


    return state
```

```python
graph = StateGraph(state_schema=AgentState)

graph.add_node("complement_user", complement_node)
graph.set_entry_point(key= "complement_user")
graph.set_finish_point(key="complement_user")


app = graph.compile()
```
