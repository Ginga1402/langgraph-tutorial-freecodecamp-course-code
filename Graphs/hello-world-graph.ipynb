```python
from typing import Dict,TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message : str

def greeing_node(state: AgentState) -> AgentState:
    """Adding simple greeting message to the state"""


    state['message'] = "Hey " + state['message'] + ", how are you doing?"

    return state
```

```python
graph = StateGraph(state_schema= AgentState)

graph.set_entry_point("create_greeting")
graph.add_node("create_greeting" , greeing_node)
graph.set_finish_point("create_greeting")


app = graph.compile()
```
